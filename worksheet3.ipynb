```python
#3.2 Exploratory Data Analysis Practice Exercise - 1.
#Which passenger had the highest fare paid relative to their age?

# Load the Titanic dataset
titanic_df = pd.read_csv('Titanic-Dataset.csv')

# Fill missing values in the Age column with the median age of the dataset
titanic_df['Age'].fillna(titanic_df['Age'].median(), inplace=True)

# Add a column fare_per_year
titanic_df['fare_per_year'] = titanic_df['Fare'] / titanic_df['Age']

# Subset rows where fare_per_year is higher than 5
high_fare_age = titanic_df[titanic_df['fare_per_year'] > 5]

# Sort high_fare_age by descending fare_per_year
high_fare_age_srt = high_fare_age.sort_values(by='fare_per_year', ascending=False)

# Select only the Name and fare_per_year columns
result = high_fare_age_srt[['Name', 'fare_per_year']]

# View the result
print(result)
```

```python
#What percent of the total fare revenue came from each passenger class?
# Calculate the total fare
total_fare = titanic_df['Fare'].sum()
print(f"Total Fare: {total_fare}")

# Total fare for first class passengers
first_class_fare = titanic_df[titanic_df['Pclass'] == 1]['Fare'].sum()
print(f"First Class Fare: {first_class_fare}")

# Total fare for second class passengers
second_class_fare = titanic_df[titanic_df['Pclass'] == 2]['Fare'].sum()
print(f"Second Class Fare: {second_class_fare}")

# Total fare for third class passengers
third_class_fare = titanic_df[titanic_df['Pclass'] == 3]['Fare'].sum()
print(f"Third Class Fare: {third_class_fare}")

# Combine fare totals into a list
fare_totals = [first_class_fare, second_class_fare, third_class_fare]

# Calculate the proportion of fare revenue by class
proportions = [fare / total_fare * 100 for fare in fare_totals]

print(f"Proportion of Fare Revenue by Class: {proportions}")
```

```python
#What percent of the total number of passengers on the Titanic belonged to each age group (e.g., child, adult, senior)?

# Fill missing values in the Age column with the median age of the dataset
titanic_df['Age'].fillna(titanic_df['Age'].median(), inplace=True)

# Create a new column age_group
def categorize_age(age):
    if age < 18:
        return 'child'
    elif 18 <= age < 65:
        return 'adult'
    else:
        return 'senior'

titanic_df['age_group'] = titanic_df['Age'].apply(categorize_age)

# Calculate the total number of passengers
total_passengers = titanic_df.shape[0]
print(f"Total Number of Passengers: {total_passengers}")

# Count the number of passengers in each age group
age_group_counts = titanic_df['age_group'].value_counts()
print(age_group_counts)

# Calculate the proportion of passengers in each age group
age_group_proportions = (age_group_counts / total_passengers) * 100

# Display the proportions as percentages
print("Proportion of Passengers in Each Age Group:")
print(age_group_proportions)
```
